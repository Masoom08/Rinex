```python
#mean squarre error
import numpy as np
y_pred=np.array([110,220,330])
y_real=np.array([100,200,300])
error=np.subtract(y_pred,y_real)
error
```

```python
sqerror=np.square(error)
sqerror
```

```python
mean_sq_error=np.mean(sqerror)
mean_sq_error
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_real)
```
