```python
# graph: line , scatter , bar , pie , histogram , density kernel

#line plot uses : regression, trend analysis, forecasting

import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,4,5,6,7])
y=np.array([22,44,55,66,77])
plt.plot(x,y,color='#ccc1d9',ls='--',lw='2',marker='*')
for i in range (len(x)):
  plt.text(x[i],y[i],(x[i],y[i]))
plt.xlim(0,10)
plt.ylim(0,100)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line graph")
plt.show()
```

```python
#two lines

import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,5)
y=np.square(x)
z=np.power(x,3)
plt.plot(x,y,color='red',ls='--',lw='2',marker='*',label='square')
plt.plot(x,z,color='g',ls=':',lw='2',marker='+' , label='cube')
plt.legend()
plt.show()
```

```python
#scatter plot : line plot with marker not line

import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10)
z=np.power(x,3)

plt.plot(x,z,color='g',ls='',lw='2',marker='o' , label='cube')
plt.legend()
plt.show()
```

```python
plt.scatter(x,z,s=100)
plt.show()
```

```python
#Bar graph
#a numpy array with name and height and a bar graph

import numpy as np
name=np.array(['A','B','C','D'])
h=np.array([145,156,160,138])
plt.bar(name,h,color=['r','g','b','k'])
plt.xlabel('name')
plt.ylabel('height')
plt.show()
```

```python
#height and weight name
import matplotlib.pyplot as plt
import numpy as np
name=np.array(['A','B','C','D'])
h=np.array([145,156,160,138])
w=np.array([40,49,60,78])
plt.bar(name,h,width=0.4,align='edge',color='r',label='height')
plt.bar(name,w,width=-0.4,align='edge',color='g',label='weight')
plt.legend()
plt.xlabel('name')
plt.ylabel('height')
plt.show()
```

```python
# density kernel

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
age= np.random.randint(5,80,30)
sns.kdeplot(age)
plt.show()
```
