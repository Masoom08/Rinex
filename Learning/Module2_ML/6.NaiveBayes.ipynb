```python
import pandas as pd 
df= pd.read_csv('play_tennis.csv')
df
```

```python
df['play'].value_counts()
```

```python
py=9/14
pn=5/14
print(py,pn)
```

```python
df.groupby(['outlook','play']).size()
```

```python
outlook= pd.crosstab(df['outlook'],df['play'])
outlook['poy']=outlook['Yes']/9
outlook['pon']=outlook['No']/9
outlook
```

```python
outlook.loc['Sunny','poy']
```

```python
temp= pd.crosstab(df['temp'],df['play'])
temp['poy']=temp['Yes']/9
temp['pon']=temp['No']/9
temp
```

```python
humidity= pd.crosstab(df['humidity'],df['play'])
humidity['poy']=humidity['Yes']/9
humidity['pon']=humidity['No']/9
humidity
```

```python
wind= pd.crosstab(df['wind'],df['play'])
wind['poy']=wind['Yes']/9
wind['pon']=wind['No']/9
wind
```

```python
x1='poy'
x2='pon'
pyes=py*outlook.loc['Sunny',x1]*temp.loc['Hot',x1]*humidity.loc['High',x1]*wind.loc['Weak',x1]
pyes
```

```python
pno=pn*outlook.loc['Sunny',x2]*temp.loc['Hot',x2]*humidity.loc['High',x2]*wind.loc['Weak',x2]
pno
```

```python
#normalised probability
yon=(pyes)/(pyes+pno)
non=(pno)/(pyes+pno)
print(non,yon)
```

```python
#Using scikit learn lib
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
#Converting the input into numerical values using encoding
from sklearn.preprocessing import OrdinalEncoder
enc= OrdinalEncoder()
```

```python
x_enc= enc.fit_transform(x)
x_enc[0]
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(x_enc,y)
```

```python
pr=df.iloc[0,:-1].values
pr=enc.transform([pr])
pr
```
