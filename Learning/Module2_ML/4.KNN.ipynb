```python
#Euclidean Distance
import pandas as pd
df=pd.read_excel('/content/Underweight-Normal.xlsx')
df
```

```python
import matplotlib.pyplot as plt
u_df=df[df['Class']=='Underweight']
n_df=df[df['Class']=='Normal']
plt.scatter(u_df['Weight'],u_df['Height'],label='Underweight')
plt.scatter(n_df['Weight'],n_df['Height'],label='Normal')
plt.scatter(57,170)
plt.legend()
plt.show()
```

```python
X=df.iloc[:,[0,1]].values
y=df.iloc[:,2].values
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
# Test data
test=[51.5,168]
neigh.predict([test])
```

```python
d,i=neigh.kneighbors([test])
```

```python
print(d)
print(i)
```
