```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['Age'],df['EstimatedSalary'])
plt.show()
```

```python
p_df=df[df['Purchased']==1]
np_df= df[df['Purchased']==0]
```

```python
plt.figure(figsize=(10,5))
plt.scatter(p_df['Age'],p_df['EstimatedSalary'],label='Purchased')
plt.scatter(np_df['Age'],np_df['EstimatedSalary'],label='Not Purchaed')
plt.legend()
plt.show()
```

```python
x=df[['Age','EstimatedSalary']].values
y=df['Purchased'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
print(x_test.shape)
print(x_train.shape)
```

```python
.67*400
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)   #data leakage
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```
