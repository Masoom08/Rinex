```python
import pandas as pd
df = pd.read_csv('/Salary_Data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['YearsExperience'],df['Salary'])
plt.title('Scatter plot')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#consider the value of input in x  and output in y
x=df.iloc[:,0].values.reshape(-1, 1)    # vector values
y=df.iloc[:,1].values
x.shape
```

```python
from sklearn.linear_model import LinearRegression   # for scikit learn x values should be 2D metrix
reg = LinearRegression().fit(x,y)
```

```python
reg.coef_ 
```

```python
reg.intercept_
```

```python
reg.predict([[2.5]])
```

```python
# Plotting the best fit line
y_pred=reg.predict(x)
plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(df['YearsExperience'],y_pred,c='r')
plt.title('Scatter Plot')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
