```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x,y= make_blobs(n_samples=100,centers=2, n_features=2,random_state=3,cluster_std=1)
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf')
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.decision_function(x)
```

```python
import numpy as np
xin = np.arange(-7,6)
yin=np.arange(-2,8)
Yin,Xin=np.meshgrid(yin,xin)
print(Yin)
print(Xin)
xy=np.vstack([Xin.ravel(),Yin.ravel()]).T
p= model.decision_function(xy).reshape(Xin.shape)
```

```python
plt.contour(Xin,Yin,p,levels=[-1,0,1],linestyles=['--','-','--'],colors='k',alpha=0.5)
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(model.support_vectors_[:,0],model.support_vectors_[:,1],s=70,facecolors='none')
plt.show()
```
