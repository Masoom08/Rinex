```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,100)
z=1/(1+np.exp(-x))
plt.plot(x,z)
plt.show()
```

```python
import tensorflow as tf
x=np.linspace(-10,10,100)
z=tf.keras.activations.sigmoid(x)
op=np.where(z>0.5,1,0)
plt.plot(x,z)
plt.scatter(x,op,c='r',s=10)
plt.show()
```

```python
# Sigmoid activation function is used at the output layer when you do binary classification
# Note: Dont use Sigmoid in Hidden layers because of Vanishing Gradient
```

```python
import numpy as np
ips=[1.3,5.1,2.2,0.7,1.1]
def softmax(ips):
  return np.exp(ips)/np.sum(np.exp(ips))

op=np.round(softmax(ips),2)
op
```

```python
import numpy as np
y_real = np.array([100, 200, 250])
y_pred1 = np.array([110,195,260]) # ANN modell
y_pred2 = np.array([105,198,255]) # ANN model2
```

```python
error=np.subtract(y_real,y_pred1)
error
```

```python
squared_error=np.square(error)
squared_error
```

```python
mean_squared_error=np.mean(squared_error)
mean_squared_error
```

```python
mse1=np.mean(np.square(np.subtract(y_real,y_pred1)))
mse1
```

```python
import tensorflow as tf
mse1=tf.keras.losses.MeanSquaredError()
mse1(y_real,y_pred1).numpy()
```

```python
mse2=tf.keras.losses.MeanSquaredError()
mse2(y_real,y_pred2).numpy()
```

```python
import numpy as np
mae1=np.mean(np.absolute(np.subtract(y_real,y_pred1)))
mae1
```

```python
import numpy as np
from math import log

def cross_entropy(a,p):
  return -((a*log(p))+((1-a)*log(1-p)))

a=[1,1,1,0,0,0]
p=[0.9,0.8,0.7,0.1,0.2,0.3]

res=[]

for i in range(len(a)):
    res.append(cross_entropy(a[i],p[i]))

mean_ce = np.mean(res)
mean_ce
```
